```python
df = df.drop(columns = ["homepage" , "original_title" , ] , axis = 1)
```

```python
dff = pd.DataFrame(columns = ["id" ,"genres" , "keywords"  , "overview"  , "title"] , data = df)
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L



```

```python
dff['genres'] = dff['genres'].apply(convert)


dff['keywords'] = dff['keywords'].apply(convert)
```

```python
dff.head(10)
```

```python
dff.isnull().sum()
```

```python
dff.dropna(inplace = True)
```

```python
dff.shape
```

```python
dff['overview'][0]
```

```python
dff['overview'] = dff['overview'].apply(lambda x:x.split())
```

```python
dff['overview'][0]
```

```python
dff['genres'] = dff['genres'].apply(lambda x:[i.replace(" ","") for i in x ])
dff['keywords'] = dff['keywords'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
dff['genres'].head()
```

```python
dff.head()
```

```python
dff['overview'][0]
```

```python
dff['tags'] = dff['overview'] + dff['genres'] + dff['keywords']
```

```python
dff.head()
```

```python
ndf = dff[['id', 'title', 'tags']]
```

```python
ndf.head()
```

```python
ndf['tags'] = ndf['tags'].apply(lambda x:" ".join(x))
```

```python
ndf['tags'] = ndf['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ndf['tags'] = ndf['tags'].apply(stem)
```

```python
import sklearn.feature_extraction.text as text
cv = text.CountVectorizer(max_features=5000, stop_words = 'english')
```

```python
ndf['tags'][0]
```

```python
vectors = cv.fit_transform(ndf['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  if(movie not in ndf['title']):
     return "Movie not found"
  else:
    index = ndf[ndf['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in distances:
        print(ndf.iloc[i[0]].title)
```

```python
recommend('tiger')
```
